```python
import pandas as pd
```

```python
# Define the file path
file_path = 'eu_bdt-ud-dev.conllu'

# Define column names
column_names = ['id', 'form', 'lemma', 'upos', '_1', '_2', '_3', '_4', '_5', '_6', '_7', '_8', '_9']

# Read the file, skipping comments and empty lines
with open(file_path, 'r', encoding='utf-8') as f:
    data = [line.strip().split('\t') for line in f if line.strip() and not line.startswith('#')]

# Pad each line with empty strings to ensure it has 13 columns
data = [row + [''] * (len(column_names) - len(row)) if len(row) < len(column_names) else row for row in data]

# Create the DataFrame
df = pd.DataFrame(data, columns=column_names)

# Convert 'id' and '_3' columns to numeric, coercing errors to NaN
df['id'] = pd.to_numeric(df['id'], errors='coerce')
df['_3'] = pd.to_numeric(df['_3'], errors='coerce')

# Print the DataFrame
print(df)

# Get a summary of the DataFrame
print("\nSummary:")
df.info()
```

```python
import matplotlib.pyplot as plt #Import Matplotlib for plotting


# Define the file path
file_path = 'eu_bdt-ud-dev.conllu'

# Define column names
column_names = ['id', 'form', 'lemma', 'upos', '_1', '_2', '_3', '_4', '_5', '_6', '_7', '_8', '_9']

# Read the file, skipping comments and empty lines
with open(file_path, 'r', encoding='utf-8') as f:
    data = [line.strip().split('\t') for line in f if line.strip() and not line.startswith('#')]


# Pad each line with empty strings to ensure it has 13 columns
data = [row + [''] * (len(column_names) - len(row)) if len(row) < len(column_names) else row for row in data]

# Create the DataFrame
df = pd.DataFrame(data, columns=column_names)


# Convert 'id' and '_3' columns to numeric, coercing errors to NaN
df['id'] = pd.to_numeric(df['id'], errors='coerce')
df['_3'] = pd.to_numeric(df['_3'], errors='coerce')

# Print the DataFrame
print(df)


# Get a summary of the DataFrame
print("\nSummary:")
df.info()


# Create a bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better visibility
plt.bar(df['id'], df['lemma']) #Bar plot: x-axis is 'id', y-axis is 'lemma'
plt.xlabel("ID")
plt.ylabel("Lemma")
plt.title("Lexical Similarity Scores") #Add a title
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.tight_layout()  # Adjust subplot parameters for a tight layout
plt.show()
```
